```python
# First, install the required packages
!pip install matplotlib pillow numpy

# Import the libraries after installation
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Method 1: Using PIL (Pillow) to load the image
# Use a relative path since the image is in the same directory as the notebook
image = Image.open('screenshot.png')  # Direct filename when in same directory

# Display the image
plt.figure(figsize=(10, 8))
plt.imshow(np.array(image))
plt.axis('off')  # Hide axes
plt.title('Loaded Screenshot')
plt.show()
```
